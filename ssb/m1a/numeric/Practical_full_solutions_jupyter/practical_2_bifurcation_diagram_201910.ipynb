
## Bifurcation diagram


Solving for 0 derivatives gives two solutions if <code>c < (1/2)</code> where <code>c = a*b</code> as <code>4c**2</code>  must be less than one.


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Set up c, and the two cases for p.

```python
c = np.arange(0.1,0.501,0.01)
p1 = np.divide(1 + np.sqrt( (1-4* np.square(c))),np.multiply(2,c))
p2 = np.divide(1 - np.sqrt( (1-4* np.square(c))),np.multiply(2,c))
```

Plot.

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(c, p1, color='tab:blue',label='p1')
ax.plot(c, p2, color='tab:orange',label='p2')
ax.set_title('Bifurcation diagram')
ax.set_xlabel('c')
ax.set_ylabel('[P]')
ax.legend()
plt.show()
```
